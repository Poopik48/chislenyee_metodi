Задание 2 : комбинированным методом хорд и касательных уточните до e=5 корень уравнения x^3+3x^2-24x+1=0

```python
def f(x):
    return x**3 + 3*x**2 - 24*x + 1

def df(x):
    return 3*x**2 + 6*x - 24

def solve():
    a = 0.0  
    b = 1.0  
    eps = 0.000001 
    
    step = 0
    while abs(b - a) > eps:
        step += 1
        
      
        f_a, f_b = f(a), f(b)
        
        #хорд
        a_new = a - (f_a * (b - a)) / (f_b - f_a)
        
        #кас
        b_new = b - f_b / df(b)
        
        #
        a, b = a_new, b_new
        
        print(f"Шаг {step}: границы стянулись до {abs(b-a):.8f}")

    print(f"\nУра найден за {step} шагов.")
    return (a + b) / 2

r = solve()
print(f"x = {r:.6f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

#4 разных случая(на 4 картинки)
funcs = [
    lambda x: (x-1.5)**2 - 0.5,           
    lambda x: -(x-1.5)**2 + 0.5,           
    lambda x: -np.sqrt(x-1) + 0.5,         
    lambda x: np.sqrt(2.5-x) - 0.5         
]

#теор смысл
titles = ["F'>0, F''>0", "F'<0, F''<0", "F'<0, F''>0", "F'>0, F''<0"]
#корни
roots = [1.5+np.sqrt(0.5), 1.5-np.sqrt(0.5), 1.25, 2.25]

#
fig, axes = plt.subplots(2, 2, figsize=(12, 10))
axes = axes.flatten()

for i, (f, title, root) in enumerate(zip(funcs, titles, roots)):
    #область определения и значения
    x = np.linspace(0.5, 2.8, 200)
    y = f(x)

    #график функ
    axes[i].plot(x, y, 'b-', linewidth=2)
    #оси координат
    axes[i].axhline(0, color='black', linewidth=0.5)
    #Отметка корня
    axes[i].axvline(root, color='g', linestyle='--', alpha=0.7, label='корень')
    #график(оформление)
    axes[i].set_title(f'Случай {chr(97+i)}: {title}')
    axes[i].grid(True, alpha=0.3)
    axes[i].legend()

plt.tight_layout()
plt.show()
```
