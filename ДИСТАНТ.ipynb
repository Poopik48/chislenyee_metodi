```python
from numpy import *
from matplotlib.pyplot import *


# задать диапозон
x = linspace(-6.5, 10, 500)
#использую np.log10 для lg
y1 = 2 * log10(x + 7)
y2 = 5 * sin(x)

# график
figure(figsize=(20, 12))
plot(x, y1, 'b-', label='2*lg(x + 7)')
plot(x, y2, 'r-', label='5*sin(x)')
grid(True)
axhline(y=0, color='k', linestyle='-', alpha=0.3)
axvline(x=0, color='k', linestyle='-', alpha=0.3)
xlabel('ось X')
ylabel('ось Y')
title('График функций')
legend()
show()

#поиск интервалов пересечения
print("Интервалы, где f(x) меняет знак:")
for i in range(len(x)-1):
    f_x1 = (2 * log10(x[i] + 7)) - (5 * sin(x[i]))
    f_x2 = (2 * log10(x[i+1] + 7)) - (5 * sin(x[i+1]))
    if f_x1 * f_x2 < 0:
        print(f"Пересечение между x = {x[i]:.2f} и x = {x[i+1]:.2f}")
```

```python
from math import *
from pandas import *

# Функция
def f(x):
    return 2 * log10(x + 7) - 5 * sin(x)


a = 0.0
b = 1.0
epsilon = 0.001
iter_count = 0

data = []

#бисекция
while (b - a) > epsilon:
    iter_count += 1
    x_mid = (a + b) / 2
    fx = f(x_mid)
    data.append([iter_count, a, b, x_mid, fx, b-a])
    
    if fx == 0:
        break
    elif f(a) * fx < 0:
        b = x_mid
    else:
        a = x_mid


x_root = (a + b) / 2

#таблица
table = DataFrame(data, columns=["Итерация", "a", "b", "x", "f(x)", "Длина интервала"])
print(table.to_string(index=False))

# Итоговые значения
print("\nКорень уравнения: {:.6f}".format(x_root))
print("Корень с точностью {}: {:.3f}".format(epsilon, x_root))
print("Количество итераций:", iter_count)
```

```python
def f(x):
    return asin(0.4 * log10(x + 7)) - x  #функция


epsilon = 0.000001  #точность

# метод бисекции
a = 0
b = 1
iter_count = 0
data_bis = []

while (b - a) / 2 > epsilon:
    iter_count += 1
    c = (a + b) / 2  #середина
    data_bis.append([iter_count, a, b, c, f(c)])

    if f(a) * f(c) < 0:  #проверка знак
        b = c
    else:
        a = c

root_bis = (a + b) / 2

table_bis = DataFrame(data_bis, columns=["Итерация", "a", "b", "c", "f(c)"])
print("МЕТОД БИСЕКЦИИ")
print(table_bis.to_string(index=False))
print("\nКорень:", round(root_bis, 6))
print("Количество итераций:", iter_count)


# метод хорд
a = 0
b = 1
iter_count = 0
data_chord = []

while abs(b - a) > epsilon:
    iter_count += 1
    x = a - f(a) * (b - a) / (f(b) - f(a))  #формула хорд
    data_chord.append([iter_count, a, b, x, f(x)])

    if f(a) * f(x) < 0:
        b = x
    else:
        a = x

root_chord = x

table_chord = DataFrame(data_chord, columns=["Итерация", "a", "b", "x", "f(x)"])
print("\n\nМЕТОД ХОРД")
print(table_chord.to_string(index=False))
print("\nКорень:", round(root_chord, 6))
print("Количество итераций:", iter_count)
```
