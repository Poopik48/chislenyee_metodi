Задание 2 : комбинированным методом хорд и касательных уточните до e=5 корень уравнения x^3+3x^2-24x+1=0

```python
def f(x):
    return x**3 + 3*x**2 - 24*x + 1

def df(x):
    return 3*x**2 + 6*x - 24

def human_solve():
    a = 0.0  
    b = 1.0  
    eps = 0.000001 
    
    step = 0
    while abs(b - a) > eps:
        step += 1
        
      
        f_a, f_b = f(a), f(b)
        
        #хорд
        a_new = a - (f_a * (b - a)) / (f_b - f_a)
        
        #кас
        b_new = b - f_b / df(b)
        
        #
        a, b = a_new, b_new
        
        print(f"Шаг {step}: границы стянулись до {abs(b-a):.8f}")

    print(f"\nУра найден за {step} шагов.")
    return (a + b) / 2

root = human_solve()
print(f"x = {root:.6f}")
```
