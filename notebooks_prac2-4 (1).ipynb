```python
from numpy import *
from matplotlib.pyplot import *


#использую np.log10 для lg
x = linspace(-6.5, 10, 500)
y1 = 2 * log10(x + 7)
y2 = 5 * sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'b-', label='2*lg(x + 7)')
plt.plot(x, y2, 'r-', label='5*sin(x)')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('ось X')
plt.ylabel('ось Y')
plt.title('Графики функций (Вариант 14)')
plt.legend()
plt.show()

#поиск интервалов пересечения
print("Интервалы, где f(x) меняет знак:")
for i in range(len(x)-1):
    f_x1 = (2 * np.log10(x[i] + 7)) - (5 * np.sin(x[i]))
    f_x2 = (2 * np.log10(x[i+1] + 7)) - (5 * np.sin(x[i+1]))
    if f_x1 * f_x2 < 0:
        print(f"Пересечение между x = {x[i]:.2f} и x = {x[i+1]:.2f}")
```

```python
import math
import numpy
import pandas

# Функция
def f(x):
    return 2 * math.log10(x + 7) - 5 * math.sin(x)


a = 0.0
b = 1.0
epsilon = 0.001
iter_count = 0

data = []

#бисекция
while (b - a) > epsilon:
    iter_count += 1
    x_mid = (a + b) / 2
    fx = f(x_mid)
    data.append([iter_count, a, b, x_mid, fx, b-a])
    
    if fx == 0:
        break
    elif f(a) * fx < 0:
        b = x_mid
    else:
        a = x_mid


x_root = (a + b) / 2

#таблица
table = pandas.DataFrame(data, columns=["Итерация", "a", "b", "x", "f(x)", "Длина интервала"])
print(table.to_string(index=False))

# Итоговые значения
print("\nКорень уравнения: {:.6f}".format(x_root))
print("Корень с точностью {}: {:.3f}".format(epsilon, x_root))
print("Количество итераций:", iter_count)
```

```python
import math

#первообразная
def g(x):
    return math.asin(0.4 * math.log10(x + 7))

x_val = 0.5
epsilon = 0.000001
iter_count = 0

data = []

#простых итераций
while True:
    iter_count += 1
    x_new = g(x_val)
    diff = abs(x_new - x_val)
    data.append([iter_count, x_val, x_new, diff])
    
    if diff < epsilon:
        x_val = x_new
        break
    x_val = x_new


x_root = x_val

#таблица
table = pandas.DataFrame(data, columns=["Итерация", "x", "g(x)", "|x - g(x)|"])
print(table.to_string(index=False))

print("\nКорень уравнения: {:.6f}".format(x_root))
print("Корень с точностью {:.6f}: {:.6f}".format(epsilon, x_root))
print("Количество итераций:", iter_count)
```
