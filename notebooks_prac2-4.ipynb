```python
import numpy as np
import matplotlib.pyplot as plt


#использую np.log10 для lg
x = np.linspace(-6.5, 10, 500)
y1 = 2 * np.log10(x + 7)
y2 = 5 * np.sin(x)

plt.figure(figsize=(10, 6))
plt.plot(x, y1, 'b-', label='2*lg(x + 7)')
plt.plot(x, y2, 'r-', label='5*sin(x)')
plt.grid(True)
plt.axhline(y=0, color='k', linestyle='-', alpha=0.3)
plt.axvline(x=0, color='k', linestyle='-', alpha=0.3)
plt.xlabel('ось X')
plt.ylabel('ось Y')
plt.title('Графики функций (Вариант 14)')
plt.legend()
plt.show()

#поиск интервалов пересечения
print("Интервалы, где f(x) меняет знак:")
for i in range(len(x)-1):
    f_x1 = (2 * np.log10(x[i] + 7)) - (5 * np.sin(x[i]))
    f_x2 = (2 * np.log10(x[i+1] + 7)) - (5 * np.sin(x[i+1]))
    if f_x1 * f_x2 < 0:
        print(f"Пересечение между x = {x[i]:.2f} и x = {x[i+1]:.2f}")
```

```python
def f(x):
    return 2 * np.log10(x + 7) - 5 * np.sin(x)

a = 0.0
b = 1.0
epsilon = 0.001
iter_count = 0

print("Итерация |     a     |     b     |     x     |    f(x)   |  Длина интервала")
print("-" * 75)

while (b - a) > epsilon:
    iter_count += 1
    x_mid = (a + b) / 2
    fx = f(x_mid)
    print(f"{iter_count:^8} | {a:^9.6f} | {b:^9.6f} | {x_mid:^9.6f} | {fx:^9.6f} | {b-a:^16.6f}")
    if fx == 0:
        break
    elif f(a) * fx < 0:
        b = x_mid
    else:
        a = x_mid

x_root = (a + b) / 2
print("-" * 75)
print(f"Корень уравнения: {x_root:.6f}")
print(f"Корень с точностью {epsilon}: {x_root:.3f}")
print(f"Количество итераций: {iter_count}")
```

```python
import math


def g(x):
    #преобразованная функция x = g(x)
    return math.asin(0.4 * math.log10(x + 7))

x_val = 0.5
epsilon = 0.000001
iter_count = 0

print("Итерация |     x     |    g(x)    |  |x - g(x)|")
print("-" * 45)

while True:
    iter_count += 1
    x_new = g(x_val)
    diff = abs(x_new - x_val)
    print(f"{iter_count:^8} | {x_val:^9.6f} | {x_new:^9.6f} | {diff:^11.6f}")
    if diff < epsilon:
        x_val = x_new
        break
    x_val = x_new

print("-" * 45)
print(f"Корень уравнения: {x_val:.6f}")
print(f"Корень с точностью 0.000001: {x_val:.6f}")
print(f"Количество итераций: {iter_count}")
```
