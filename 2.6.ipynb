Задание 1: Метод половинного деления трансцидентное уравнение lgx−cosx=0 точность е=4

```python
import math
import matplotlib.pyplot as plt
import numpy as np

def f(x):
    return math.log10(x) - math.cos(x)

def bisection(a, b, eps=1e-4):
    if f(a) * f(b) > 0:
        print("нет сменв знака на отрезке")
        
    while (b - a) / 2 > eps:
        c = (a + b) / 2
        if f(a) * f(c) <= 0:
            b = c
        else:
            a = c
            
    return (a + b) / 2

r = bisection(1, 2)
print("корень:", round(r, 4))


x = np.linspace(0.5, 2.5, 100)
y = [f(i) for i in x]
plt.plot(x, y)
plt.axhline(0, color='red', linestyle='--')
plt.scatter(r, 0, color='black')
plt.grid()
plt.show()
```

2 задание : Любым итерационным методом(простой итерации) найти корень уравнения x²-3,2x=1 с точностью е=4

```python
def phi(x):
    return (x**2 - 1) / 3.2

def simple_iteration(x0, eps=1e-4):
    while True:
        x1 = phi(x0)
        if abs(x1 - x0) < eps:
            return x1
        x0 = x1

r = simple_iteration(1)
print("корень:", round(r, 4))


x = np.linspace(-1, 4, 100)
y = [i**2 - 3.2*i - 1 for i in x]
plt.plot(x, y)
plt.axhline(0, color='red', linestyle='--')
plt.scatter(r, 0, color='black')
plt.grid()
plt.show()
```

Задание 3 : найти наибольший корень уравнения 4x-3lnx=4 с точностью е=4 методом итерации корни отделить графически

```python
def g(x):
    return (4 + 3 * math.log(x)) / 4

def iteration(x0, eps=1e-4):
    while True:
        x1 = g(x0)
        if abs(x1 - x0) < eps:
            return x1
        x0 = x1

r = iteration(2)
print("наибольший корень:", round(r, 4))


x = np.linspace(0.1, 4, 100)
y = [4*i - 3*math.log(i) - 4 for i in x]
plt.plot(x, y)
plt.axhline(0, color='red', linestyle='--')
plt.scatter(r, 0, color='black')
plt.grid()
plt.show()
```
